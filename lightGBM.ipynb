```python
trains = lgb.Dataset(x_train, y_train)
valids = lgb.Dataset(x_test, y_test)
```

```python
params = {
    'boosting_type': 'gbdt',
    'num_leaves': 4000,
    "objective": "regression_l1",
    "metrics": "mae",
    'feature_pre_filter': False,
    'learning_rate': 0.05,
    'bagging_fraction': 1.0,
    'feature_fraction': 0.4,
    'random_state': 0,
    'lambda_l1' : 6.566294695479466,
    'lambda_l2' : 5.925006875927771, 
    'bagging_freq': 0,
    'min_child_samples': 25,
    }

model = lgb.train(params,
                  trains, 
                  valid_sets=valids, 
                  num_boost_round=10000, 
                  early_stopping_rounds=500,
                  categorical_feature = cat_list,
                  verbose_eval=200,
                  )
```

```python
# 特徴量の重要度グラフ
lgb.plot_importance(model, height = 0.5, figsize = (5,10), importance_type='gain')

train_pred = model.predict(x_train, num_iteration=model.best_iteration)
eval_pred = model.predict(x_test, num_iteration=model.best_iteration)
predict = model.predict(test, num_iteration=model.best_iteration)
```

```python
#  r2_score を計算
from sklearn.metrics import mean_absolute_error
train_score = mean_absolute_error(y_train, train_pred)
eval_score = mean_absolute_error(y_test, eval_pred)
print(train_score)
print(eval_score)
```

```python
# 時間計測開始
time_sta = time.time()

# Optuna
import optuna.integration.lightgbm as lgb_o
trains_o = lgb_o.Dataset(x_train, y_train)
valids_o = lgb_o.Dataset(x_test, y_test,reference=trains_o)

# ハイパーパラメータサーチ&モデル構築
params = {'objective': 'regression_l1',
          'metric': 'mae',
          'random_seed':0
}

model_o = lgb_o.train(params,
                  trains_o, 
                  valid_sets=valids_o, 
                  early_stopping_rounds=200,
                  categorical_feature = cat_list,
                  verbose_eval=200,
                  )

# ベストパラメータの取得
best_params = model_o.params
print("  Params: ")
for key, value in best_params.items():
    print("    {}: {}".format(key, value))

# 時間計測終了
time_end = time.time()
# 経過時間（秒）
tim = time_end- time_sta
print(tim)
```
